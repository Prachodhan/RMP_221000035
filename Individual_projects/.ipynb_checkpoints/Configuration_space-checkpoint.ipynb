```python
import matplotlib.pyplot as plt
import numpy as np

# Workspace size
xlim = (0, 20)
ylim = (0, 20)

# Obstacles (as circles): (x, y, radius)
obstacles = [
    (5, 5, 2),
    (12, 8, 3),
    (15, 15, 2.5),
    (7, 15, 2),
]

# Create grid
resolution = 0.1
x = np.arange(xlim[0], xlim[1], resolution)
y = np.arange(ylim[0], ylim[1], resolution)
X, Y = np.meshgrid(x, y)

# Initialize configuration space (0: free, 1: obstacle)
cspace = np.zeros_like(X)

# Populate C-space
for (ox, oy, r) in obstacles:
    distance = np.sqrt((X - ox)**2 + (Y - oy)**2)
    cspace[distance <= r] = 1  # Mark obstacle regions

# Plotting
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_title("Configuration Space (C-Space) of a Point Robot")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_xlim(xlim)
ax.set_ylim(ylim)

# Show occupied space
ax.contourf(X, Y, cspace, levels=[0.5, 1], colors='gray', alpha=0.7)

# Overlay obstacle circles for clarity
for (ox, oy, r) in obstacles:
    circle = plt.Circle((ox, oy), r, color='black', fill=False, linestyle='--')
    ax.add_patch(circle)

plt.grid(True)
plt.gca().set_aspect('equal')
plt.show()
```
